# SQL DATABASE

```python
import sqlite3
import pandas as pd

# Connect to one database file (usually better than 4 separate files)
conn = sqlite3.connect("Cross_Market_Analysis")
cursor = conn.cursor()

# Define specific schemas for your 4 tables
tables = {
    "Cryptocurrencies": "id VARCHAR(50) PRIMARY KEY, symbol VARCHAR(10),	name VARCHAR(100),	current_price DECIMAL(18, 6),	market_cap BIGINT,	market_cap_rank INT,	total_volume BIGINT,	circulating_supply DECIMAL(20, 6),	total_supply DECIMAL(20, 6),	ath DECIMAL(18, 6),	atl DECIMAL(18, 6),	date DATE",
    "Crypto_prices": "coin_id	VARCHAR(50), date	DATE, prices_usd DECIMAL(18, 6),FOREIGN KEY (coin_id) REFERENCES Cryptocurrencies(id)",
    "Oil_prices": "Date DATE PRIMARY KEY, Price DECIMAL(18, 6)",
    "Stocks_prices": "Date DATE,	Open DECIMAL(18, 6),	High DECIMAL(18, 6) , Low DECIMAL(18, 6),	Close DECIMAL(18, 6),	Volume DECIMAL(18, 6),	Ticker DECIMAL(18, 6)"
}

for table_name, schema in tables.items():
    cursor.execute(f"CREATE TABLE IF NOT EXISTS {table_name} ({schema})")
```

```python
# List of pairs
df_list = [
    (crpyto_curr, "Cryptocurrencies"),
    (crpyto_price, "Crypto_prices"),
    (oil_price, "Oil_prices"),
    (stocks_price, "Stocks_prices")
]

for df, table_name in df_list:
    # FIX: Wrap column names in [] to handle colons or spaces
    cols = ", ".join([f"[{col}]" for col in df.columns])
    placeholders = ", ".join(["?"] * len(df.columns))

    # FIX: Wrap table name in [] just in case
    sql = f"INSERT OR REPLACE INTO [{table_name}] ({cols}) VALUES ({placeholders})"

    for _, row in df.iterrows():
        # Ensure values are passed as a clean tuple
        cursor.execute(sql, tuple(row))

conn.commit()

print("Data inserted successfully without token errors.")
```
